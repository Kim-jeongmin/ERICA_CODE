```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.binomial(n=1, p=0.5, size=1000)

x = []
index = 0
while(index != 1000):
  x.append(index)
  index+=1

coin_cnt = 0
index = 0

y = []
while(index != 1000):
  if data[index] == 0: coin_cnt+=1

  if index == 0 : y.append(coin_cnt)
  else : y.append(coin_cnt/(index+1))
  index+=1




plt.title('Coin : p(H)=' + str(round(sum(data)/1000,4)))
plt.grid()


plt.plot(x, y, linestyle = '-')
plt.show()


```
